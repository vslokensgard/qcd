```python
import numpy as np
import matplotlib.pyplot as plt
import visual as vs
import pionkaon as pk
from scipy import optimize as opt
```

```python
def jacknifeGraph(eff_energies, jack_error, err_cutoff=None, selected_bins=None):
    energies_by_bin = np.transpose(eff_energies)
    num_bins, num_intervals = energies_by_bin.shape
    min_slice = 1
    if err_cutoff:
        max_slice = pk.maxSlice(jack_error, err_cutoff)
    else:
        max_slice = num_intervals

    x = range(min_slice, max_slice)
    jack_error = jack_error[min_slice:max_slice]

    if selected_bins:
        to_plot = selected_bins
    else:
        to_plot = list(range(num_bins))

    for bin_no in to_plot:
        y = energies_by_bin[bin_no][min_slice:max_slice]
        plt.scatter(x, y)
        plt.errorbar(x, y, yerr=jack_error, fmt="o")
        plt.show()

    return

momentum_5 = np.loadtxt("./export/momentum5")
energy_5 = pk.effectiveEnergy(momentum_5)
energy_5_error = pk.jacknifeError2D(energy_5)
selected_bins = range(0, len(energy_5_error), 10)
jacknifeGraph(energy_5, energy_5_error, err_cutoff=100, selected_bins=selected_bins)
```

```python
import numpy as np

# Purpose:    plot and quadratic-fit a set of momenta with their average energies and error
# Parameters: 2D NumPy array VALS_BY_STATE returned by listGroundStates()
#             string SAVE_AS, path and file name where the dispersion relation plot will be saved
# Returns:    void, dispersion relation plot and fit function saved to an external file
def plotQuad(vals_by_state, save_as=None):
    states = len(vals_by_state)
    x = np.empty(states)
    y = np.empty(states)
    e = np.empty(states)
    if not vals_by_state:
        return None
    for i in range(len(vals_by_state)):
        x[i], y[i], e[i] = vals_by_state[i]

    fit = np.polyfit(x, y, 2)
    fit_func = np.poly1d(fit)
    polyline = np.linspace(0, len(x), 100)
    plt.scatter(x, y)
    e *= 2
    plt.errorbar(x, y, yerr=e, fmt="o")
    plt.plot(polyline, fit_func(polyline))
    plt.xlabel(str(fit_func))

    if save_as: plt.savefig(save_as)
    plt.show()
    return

# Purpose:    plot and quadratic-fit a set of momenta with their average energies and error
# Parameters: 2D NumPy array VALS_BY_STATE returned by listGroundStates()
#             string SAVE_AS, path and file name where the dispersion relation plot will be saved
# Returns:    void, dispersion relation plot and fit function saved to an external file
def plotLin(vals_by_state, save_as=None):
    states = len(vals_by_state)
    x = np.empty(states)
    y = np.empty(states)
    e = np.empty(states)
    for i in range(len(vals_by_state)):
        x[i], y[i], e[i] = vals_by_state[i]

    fit = np.polyfit(x, y, 1)
    fit_func = np.poly1d(fit)
    polyline = np.linspace(0, 1, 100)
    plt.scatter(x, y)
    plt.errorbar(x, y, yerr=e, fmt="o")
    plt.plot(polyline, fit_func(polyline))
    plt.xlabel(str(fit_func))

    if save_as: plt.savefig(save_as)
    plt.show()
    return

def scaleXAxis(vals_by_state):
    vals_by_axis = np.transpose(vals_by_state)
    vals_by_axis[0] *= (np.pi * 2) / 32
    vals_by_state = np.transpose(vals_by_axis)
    return vals_by_state
    
results_squared = np.square(results)
scaled_results = scaleXAxis(results)
plotLin(np.square(scaled_results))
```
